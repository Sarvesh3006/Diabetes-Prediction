# Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
# loading data in notebook
diabetes_data = pd.read_csv("Diabetes.csv")
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
# Getting the statistical measure of data
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

* Label 0 = non diabetic patients
* Label 1 = Diabetic patients

```python
diabetes_data.groupby('Outcome').mean()
```

```python
# Saperating the data
x = diabetes_data.drop(columns = "Outcome", axis = 1)
y = diabetes_data['Outcome']
```

```python
x
```

```python
y
```

# Standardization of data

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
y = diabetes_data["Outcome"]
```

```python
x,y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 42)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
clf = svm.SVC(kernel = 'linear')
```

```python
# training the support vector machine classifier
clf.fit(x_train, y_train)
```

# Evaluating our model

* Finding Accuracy score

```python
# accuracy score on training data
x_train_preds = clf.predict(x_train)
training_data_accuracy = accuracy_score(x_train_preds, y_train)
print(f"accuracy score on training data is : {training_data_accuracy:.2f}%")
```

```python
# accuracy score on test data
x_test_preds = clf.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_preds, y_test)
print(f"accuracy score on test data is : {testing_data_accuracy:.2f}%")
```

# Making the predictive system

```python
input_data = (1,85,66,29,0,26.6,0.351,31)

# changing input data to numpy array
np_input_data = np.asarray(input_data)

#reshaping the data
input_data_reshape = np_input_data.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshape)

#predicting the outcome
pred = clf.predict(std_data)
print(pred)

if (pred[0] == 0 ):
    print('the person is non diabetic')
else:
    print('the person is diabetic')
```
